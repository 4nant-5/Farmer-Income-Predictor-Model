```python
num_features = df.select_dtypes(include=['int64', 'float64']).columns.drop(['income', 'FarmerID'])
cat_features = df.select_dtypes(include=['object', 'category']).columns
```

```python
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
preprocessor = ColumnTransformer([
    ('num', num_pipeline, num_features),
    ('cat', cat_pipeline, cat_features)
])
```

```python
model_pipeline = Pipeline([
    ('preprocessing', preprocessor),
    ('model', RandomForestRegressor(n_estimators=100, random_state=42))
])
```

```python
X_train = df.drop(columns=['income', 'FarmerID'])
y_train = df['income']
model_pipeline.fit(X_train, y_train)
```

```python
test_df = pd.read_csv("test data.csv")
```

```python
y_train_pred = model_pipeline.predict(X_train)
mae = mean_absolute_error(y_train, y_train_pred)
rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
r2 = r2_score(y_train, y_train_pred)

print(f"Train MAE:  {mae:,.2f}")
print(f"Train RMSE: {rmse:,.2f}")
print(f"Train R²:   {r2:.3f}")
```

```python
X_test = test_df.drop(columns=["FarmerID"])
```

```python
test_predictions = model_pipeline.predict(X_test)
```

```python
submission = pd.DataFrame({
    "farmer_id": test_df["FarmerID"],
    "predicted_income": test_predictions
})
```

```python
submission.head()
```

```python
submission.to_csv("final_predictions.csv", index=False)
```

```python
def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    non_zero = y_true != 0
    return np.mean(np.abs((y_true[non_zero] - y_pred[non_zero]) / y_true[non_zero])) * 100

mape = mean_absolute_percentage_error(y_train, y_train_pred)
print(f"Train MAPE: {mape:.2f}%")
```
